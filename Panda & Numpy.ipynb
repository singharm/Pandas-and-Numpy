```python
customer_id=["100","101","102","103","104"]
sales_1=[230,450,80,500,100]
sales_2=[180,630,70,600,110]
sales_3=[170,550,60,700,90]
description=["male,35,full time employee,first criteria:quality", "female,40,manager, interested in new features",
             "male, 22, student, price sensitive", " female, 75, retired, interested in new features","female, 25, part-time, price sensitive"]
value=["medium", "high", "low", "high", "low" ]
transaction=pd.DataFrame({"customer_id": customer_id, "sales_1": sales_1, "sales_2": sales_2, "sales_3": sales_3, "description": description, "customer_value": value})
transaction
```

```python
transaction["total_sales"]=transaction["sales_1"]+transaction["sales_2"]+transaction["sales_3"]
```

```python
total=transaction[["sales_1","sales_2","sales_3"]].apply(np.sum)
```

```python
total
```

```python
average=transaction[["sales_1","sales_2","sales_3"]].apply(np.mean, axis = 0)
```

```python
average
```

```python
average_sales_customer=transaction[['sales_1', 'sales_2', 'sales_3']].apply(np.mean, axis=1)
average_sales_customer
```
